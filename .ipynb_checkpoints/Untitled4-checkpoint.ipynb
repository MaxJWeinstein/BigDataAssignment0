# Intro to Python

## Arithmetic

Let's begin with simple mathematical operations. We will start with two variables: num_1 and num_2. We set them to 1 and 2 respectively.

```python
num_1 = 1
num_2 = 2
```

We can set new variables equal to expressions containing old ones. The next line serves as a basic example.

You can also multiply, divide, square, log, and even raise numbers to an exponent. In the last few cases. we will use a package called numpy to assist us.

```python
import numpy as np

# Addition
num_3 = num_1 + num_2

# Multiplication
num_3 = num_3*num_2

# Divide
num_3 = num_2/num_1

# Square: num_3^num_2
num_3 = num_3**num_2

# Log
num_3 = np.log(num_2)

# Exponent
num_3 = np.exp(num_3)
```

Now that you have a basic understanding of math operations, let's briefly go over how we evaluate the "truthiness" of the statement. First, we'll reset the value of num_3 to num_1 + num_2.

```python
num_3 = num_1 + num_2
```

```python
print(f'{num_1} + {num_2} = {num_1 + num_2}')
```

## Booleans and If Statements

Then we'll evaluate the truth of a statement using the "==" operator; this is a completely different tool from "=" because "==" checks if two variables are equal; it does not reset the values of any variables. "=" simply sets the value of one variable to another.

```python
num_3 == num_1 + num_2
```

```python
num_1 == num_2
```

```python
if num_1 == num_2:
    print('1 is equal to 2')
else:
    print('1 is not equal to 2')
```

We can also use "elif" statements if we want to have more than two possible reactions to a given problem.

```python
if num_1 > num_2:
    print('1 is greater than 2')
elif num_1 == num_2:
    print('1 is equal to 2')
else:
     print('1 is less than 2')
```

You can get more creative with elif statements. Observe.

```python
name = input("What is your name")

if name == 'Nobody':
    print('Hello Nobody')
elif name == 'Somebody':
    print('Somebody is here')
elif name == 'Dobler':
    print('praise dobler')
else:
    print(f'whats good {name}')
```

What if you want to check multiple conditions at the same time? Use and/or statements.

```python
name = input("What is your name")

if 'dobler' in name or 'Dobler' in name:
    print('praise dobler')
elif 'a' in name and 'e' in name:
    print('It looks like your name contains very popular vowels.')
elif 'x' in name:
    print('Your name must be very interesting.')
elif 'q' in name and not 'u' in name:
    print('is your name qatar by chance')
else:
    print(f'whats good {name}')
```

IMPORTANT NOTE: elif statements mean that if two statements are both true, only the first one mentioned by an elif block is run. All later statements are ignored. Here is why you should be aware of this.

## Loops

```python
name = 'Beatrice Q. Dobler'

if 'dobler' in name or 'Dobler' in name:
    print('praise dobler')
elif 'a' in name and 'e' in name:
    print('It looks like your name contains very popular vowels.')
elif 'x' in name:
    print('Your name must be very interesting.')
elif 'q' in name and not 'u' in name:
    print('is your name qatar by chance')
else:
    print(f'whats good {name}')
```

```python
name = 'Beatrice Q. Dobler'

if 'dobler' in name or 'Dobler' in name:
    print('praise dobler')
if 'a' in name and 'e' in name:
    print('It looks like your name contains very popular vowels.')
if 'x' in name:
    print('Your name must be very interesting.')

# Note that the else statement is no longer tied to the three if statements above. This is important to know for later.

if ('q' in name or 'Q' in name) and not 'u' in name:
    print('is your name qatar by chance')
else:
    print(f'whats good {name}')
```

As you can see, if you just have "if" statements, every if statement will be checked separately. 

Make sure you know how else statements are linked to if statements. Test different combinations to solidify your understanding.

Now we will cover loops. The most common type of loops you'll use are for loops. Here's an example.

```python
for i in range(10):
    print(i)
```

Let's say we want to break out of the loop early, for whatever reason. We can do that too.

```python
for i in range(10):
    if i == 7:
        break
    print(i)
```

Great. Now the loop ends at 6. ut what if we just wanna SKIP a number instead of ending entirely?

```python
for i in range(10):
    if i == 7:
        continue
    print(i)
```

## Functions

Sweet. Now let's say there's an error during one of the steps in our code. For example, the case below.

```python
nums = [0, 4, 2, 8, 4, 6, None, 2, 5, 1, 73]

for num in nums:
    print(f'{num} + 22 = {num + 22}')
```

As you can see, everything's fine till we reach the None value in our code. Here's how we can skip past the step and continue as if nothing happened:

```python
nums = [0, 4, 2, 8, 4, 6, None, 2, 5, 1, 73]

for num in nums:
    try:
        print(f'{num} + 22 = {num + 22}')
    except Exception as e:
        print(f'Error for {num}: {e}')
```

We will also briefly go over functions; these are effectively black boxes that take in variable inputs and release a distinct output. Observe:

```python
def add_and_multiply(num_1, num_2, num_3):
    return (num_1 + num_2)*num_3

num_1 = 1
num_2 = 2
num_3 = 3

num_4 = add_and_multiply(num_1, num_2, num_3)

print(num_4)
```

```python
for num_1 in range(10):
    num_4 = add_and_multiply(num_1, num_2, num_3)
    print(num_4)
```

Note the return statement; it works by instructing the function to terminate and return whatever variable it has at the point of termination. 
Here's an example that integrates all the stuff we have learned so far:

```python
def find_max(arr):
    max_item = 0
    for item in arr:
        try:
            if item > max_item:
                max_item = item
        except:
            continue
    return max_item
            
arr = [1, 34, 12, 'squilliam', None, 0, -100, 2003]

max_item = find_max(arr)
print(max_item)
```

The above works for the first case, but notice if I input None for the array:

## MatPlotLib

```python
arr = None

max_item = find_max(arr)
print(max_item)
```

```python
def find_max(arr):
    max_item = 0
    if arr is None:
        return max_item
    for item in arr:
        try:
            if item > max_item:
                max_item = item
        except:
            continue
    return max_item

# First case
            
arr = [1, 34, 12, 'squilliam', None, 0, -100, 2003]

max_item = find_max(arr)
print(max_item)

# Second case

arr = None

max_item = find_max(arr)
print(max_item)
```

Finally we can focus on matplotlib. This is a popular graphing tool that can be used for pretty much anything. In our case we will use it to make a line.

```python
import matplotlib.pyplot as plt

x = np.arange(0, 10)
y = np.arange(0, 10)

plt.scatter(x, y, color='orange')
plt.plot(x, y, color='black')
plt.show()
```

We can also make more interesting plots such as this parabola:

```python
x = np.arange(-12, 12)
y = x**2

plt.plot(x, y, color='black')
plt.show()
```

```python
x = np.arange(-12, 12)
y = x**2

plt.plot(x, y, color='black')
plt.xlabel('hour of the day')
plt.ylabel('swag indicator')
plt.show()
```

Say those ticks on the sides of the graph are crampig your style. You can easily remove them

```python
x = np.arange(-12, 12)
y = x**2

plt.plot(x, y, color='black')
plt.xlabel('hour of the day')
plt.ylabel('swag indicator')
plt.tick_params(left=False, bottom=False)
plt.show()
```

You may even remove the actual numbers from the graph if you wish.

```python
x = np.arange(-12, 12)
y = x**2

plt.plot(x, y, color='black')
plt.xlabel('hour of the day')
plt.ylabel('swag indicator')
plt.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
plt.show()
```

Here's an example of a rather detailed graph you can make using matplotlib that includes titles, a legend, and demarcation lines

```python
x = np.arange(-12, 12)
y = np.sin(x)

plt.plot(x, y, color='black')
plt.xlabel('hour of the day')
plt.ylabel('status report')
plt.tick_params(bottom=False, labelbottom=False)
plt.yticks(ticks=[-1, 0, 1], labels=['its so over', 'bros?', 'were so back'])
# Horizontal lines
plt.axhline(y=1, color='green', linestyle='--')
plt.axhline(y=0, color='blue', linestyle='--')
plt.axhline(y=-1, color='red', linestyle='--')

# We can reset boundaries of our graph too!
plt.ylim(-2, 2)

plt.title('a day in the life')
plt.show()
```

Here are some last bits of advice. Say you want to put two graphs in one plot. You can do that like so:

```python
fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')

x = np.linspace(0, 24, 240)
y1 = np.sin(x)
y2 = (np.sin(x)**2)/(1 - np.sin(x))

axs[0].plot(x, y1)
axs[1].plot(x, y2)

plt.show()
```

```python
fig, axs = plt.subplots(2, 2)
fig.suptitle('Quad style subplots')

x = np.linspace(0, 24, 240)
y1 = np.sin(x)
y2 = (np.sin(x)**2)/(1 - np.sin(x))
y3 = x**2
y4 = np.exp(-0.1*(x-12)**2)

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x, y4)

plt.show()
```

Everything looks a little cluttered so let's clean that up, too.

```python
fig, axs = plt.subplots(2, 2)
fig.suptitle('Quad style subplots')

x = np.linspace(0, 24, 240)
y1 = np.sin(x)
y2 = (np.sin(x)**2)/(1 - np.sin(x))
y3 = x**2
y4 = np.exp(-0.1*(x-12)**2)

axs[0, 0].tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
axs[0, 0].plot(x, y1)
axs[0, 1].tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
axs[0, 1].plot(x, y2)
axs[1, 0].tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
axs[1, 0].plot(x, y3)
axs[1, 1].tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
axs[1, 1].plot(x, y4)

plt.show()
```

You can also add other things like colorbars! Here's a final example using a package called "pcolormesh."

```python
data = np.random.random((100, 100))

plt.pcolormesh(data)
plt.show()
```

As you can see, pcolormesh takes an array and presents it like an image. This looks rather ugly and uninteresting so let's spice it up.

```python
import scipy.ndimage

data = np.random.random((100, 100))
blurred_data = scipy.ndimage.gaussian_filter(data, sigma=25)

plt.pcolormesh(blurred_data, cmap='seismic')
plt.show()
```

Looks better already. But why don't we add a colorbar and axes just to get an idea of what to do with this plotting style?

```python
data = np.random.random((100, 100))
blurred_data = scipy.ndimage.gaussian_filter(data, sigma=10)

plt.title('Power Signal as a function of frequency and time')
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency (Hz)')
plt.pcolormesh(blurred_data, cmap='seismic', vmin=0.45, vmax=0.55)
plt.colorbar()
plt.show()
```
